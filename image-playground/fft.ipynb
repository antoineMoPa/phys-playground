```python
% matplotlib inline
import matplotlib.image as npimg
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import math
```

```python
def show(image):
    plt.imshow(image,cmap=cm.Greys)
    plt.show()
```

```python
img = np.asarray(npimg.imread("images/image-1.jpg"))
img = img.astype(float)

img = img/np.max(img)
```

```python
show(img)
# Photo by Sean Lamb CCBY-SA 2.0
```

```python
img = np.fft.fft(img,axis=0)
```

```python
plt.plot(img[0,:])
plt.plot(-1+np.exp(np.arange(0,700)/img.shape[0]))
```

```python
r = np.real(img)
i = np.imag(img)
```

```python
w = img.shape[0]
```

```python
img = r + (i*(np.exp(r/w)-1)) * 1j
```

```python
img = np.fft.ifft(img,axis=0)
```

```python
img = np.real(img)
```

```python
show(img)
```
