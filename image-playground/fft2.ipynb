# Fun with np.fft2

```python
% matplotlib inline
import matplotlib.image as npimg
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import math
```

```python
def show(image):
    plt.imshow(image,cmap=cm.Greys)
    plt.show()
```

```python
img = np.asarray(npimg.imread("images/image-1.jpg"))
img = img.astype(float)

original = img = img/np.max(img)
```

```python
show(img)
# Photo by Sean Lamb CCBY-SA 2.0
```

```python
img = np.fft.fft2(img)
```

```python
plt.plot(np.real(img[0,:]))
```

```python
r = np.real(img)
i = np.imag(img)
```

```python
w = img.shape[0]
h = img.shape[1]
```

```python
img = (r + i * 1j) * np.exp(1j*1 * r/w)
```

```python
show(np.real(img))
```

```python
img = np.fft.ifft2(img)
```

```python
img = np.real(img)
```

```python
show(img)
```

```python
img = r + 1j*i
img = -2 * img + np.roll(img,1,0) - np.roll(img,-1,0)
inv = np.fft.ifft2(img)
real = np.real(inv)
show(real)
```

```python
img = r + 1j*i
img = np.roll(img,-1,0) - np.roll(img,1,0) + np.roll(img,-1,1) - np.roll(img,1,1)
inv = np.fft.ifft2(img)
real = np.real(inv)
show(real/4)
```
